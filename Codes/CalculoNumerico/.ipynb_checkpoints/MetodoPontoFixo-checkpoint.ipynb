```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def f(x):
    return (np.power(x,2) + x - 6)

def phi1(x):
    return (6 - np.power(x,2))

def phi2(x):
    return (np.sqrt(6 - x))
```

### É possível escrever a função de iteração $\phi(x) = x$ da equaçao $x^2 + x - 6 = 0$ de duas formas:

\begin{equation}
    \phi_1(x) = x = 6 - x^2
\end{equation}

\begin{equation}
    \phi_2(x) = x = (6 - x)^{1/2}
\end{equation}

### Vamos visualizar graficamente ambas as funções de iteração junto com a função $f(x)$.

```python
x = np.arange(-4.0, 4.0, 0.01)
plt.figure()
plt.subplot(211)
plt.plot(x, f(x), "-b", label = "$f(x)$")
plt.plot(x, phi1(x), "r--", label = "$\phi1(x)$")
plt.grid()
plt.legend()
plt.show()

plt.figure()
plt.subplot(212)
plt.plot(x, f(x), "-b", label = "$f(x)$")
plt.plot(x, phi2(x), "m--", label = "$\phi2(x)$")
plt.grid()
plt.legend()
plt.show()
```

## =========================== Número de Iterações ===========================

```python
# Testando a iteração para a função phi1(x)

x_inicial = 1.5
valoresIteracao1 = []

for i in range(0, 10):
    valoresIteracao1.append(x_inicial)
    x_inicial = phi1(x_inicial)

print(valoresIteracao1) # Veremos a divergência
```

```python
# Testando a iteração para a função phi2(x)

x_inicial = 1.5
valoresIteracao2 = []

for i in range(0, 10):
    valoresIteracao2.append(x_inicial)
    x_inicial = phi2(x_inicial)

print(valoresIteracao2) # Veremos a convergência para x aproximadamente igual a 2
```

### Vamos testar o exemplo de aula: $e^x - 4x = 0$. Esta equação possui uma raiz no intervalo $[0, 1]$ e outra no intervalo $[2, 3]$. Veremos que a função de iteração $\phi(x)$ deverá ser diferente em cada um dos intervalos.

```python
def func(x):
    return ((math.e)**x/4)

def primeiroPhi(x):
    return ((math.e)**x/4)

def segundoPhi(x, p = 0.1):
    return (x + p*(4*x - (math.e)**x))
```
